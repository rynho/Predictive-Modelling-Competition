# Sample code for calculate correlation between stock and index.

```python
import numpy as np
import pandas as pd
pd.options.display.max_rows = 10

import matplotlib.pyplot as plt
%matplotlib inline

import pandas_datareader.data as web
import datetime
```

```python
start = datetime.datetime(2018, 6, 1)
end = datetime.datetime(2018, 7, 1)
```

```python
aapl = web.DataReader('AAPL', 'morningstar', start, end)
msft = web.DataReader('MSFT', 'morningstar', start, end)

df = pd.DataFrame(aapl.reset_index(level=0).Close)
df = df.rename(columns={'Close':'aapl'})
df['msft']=msft.reset_index(level=0).Close
df
```

```python
sp500 = web.DataReader('SPX', 'morningstar', start, end)
sp500
```

```python
b0 = b1 = 1
matrix = pd.DataFrame(b0*df.aapl + b1*df.msft,columns=['comb'])
matrix['sp500']=sp500.reset_index(level=0).Close

matrix.corr()
```
